```python
import pandas as pd
from sqlalchemy import create_engine


db_user = 'root'       
db_password = 'root'   
db_host = 'DESKTOP-FHJBTK2'           
db_port = '3306'                
db_name = 'olist' 


engine = create_engine(f'mysql+pymysql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')


orders_df = pd.read_csv(r"C:\Users\prashu\Downloads\archive (2)\olist_orders_dataset.csv")
order_items_df = pd.read_csv(r"C:\Users\prashu\Downloads\archive (2)\olist_order_items_dataset.csv")
products_df = pd.read_csv(r"C:\Users\prashu\Downloads\archive (2)\olist_products_dataset.csv")
sellers_df = pd.read_csv(r"C:\Users\prashu\Downloads\archive (2)\olist_sellers_dataset.csv")


display(orders_df.head())
display(order_items_df.head())



orders_df.to_sql(name='orders', con=engine, if_exists='append', index=False)
order_items_df.to_sql(name='order_items', con=engine, if_exists='append', index=False)
products_df.to_sql(name='products', con=engine, if_exists='append', index=False)
sellers_df.to_sql(name='sellers', con=engine, if_exists='append', index=False)

print("✅ All data successfully imported into MySQL.")
```
